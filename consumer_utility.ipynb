### Consumer Utility
Consumer utility refers to the level of satisfaction or happiness a consumer receives from consuming a good or service. In economics, the concept of consumer utility is used to help understand consumer behavior and make predictions about how consumers will respond to changes in prices or income.Consumer utility is often measured using a utility function, which assigns a numerical value to each possible combination of goods that reflects the consumer's level of satisfaction with that combination.

For example, the utility function is $u(X,Y) = \sqrt(X \cdot Y)$, which means that the consumer's level of satisfaction with a particular combination of goods x and y is given by the square root of their product. The square root function is a common utility function for two goods, as it exhibits diminishing marginal utility, meaning that as the consumer consumes more of one good while holding the other constant, the additional satisfaction derived from each additional unit of the good decreases. The graph below show this utility funtion in 3-D.




```python
import matplotlib
from matplotlib import animation
import numpy as np
import matplotlib.pyplot as plt

# define utility function
def get_utility(x,y):
    return np.sqrt(x*y)

fig = plt.figure(figsize=(12,8))  #set the default figure size
ax = fig.add_subplot(1,1,1, projection='3d') # add 3rd dimension to subplots

x = np.linspace(0,100,1000)
y = np.linspace(0,100,1000)
X, Y = np.meshgrid(x,y) # create a numpy 2d array
Z = get_utility(X,Y)

# graph a 3d surface plot
ax.plot([0,0,0], [100,100,100])
ax.plot_surface(X,Y,Z, alpha=0.75, color='lightblue')
ax.set_facecolor('y')
fig.set_facecolor('y')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('utility')
plt.title('Utility Function')

# graph a contour plot on z direction to show isoquants
cset = ax.contour(X,Y,Z, zdir='z', offset=0, cmap=matplotlib.cm.coolwarm)
#ax.set_xlim3d(0,60)

ax.view_init(30,-60)

```


#### Indifference Curves:

Indifference curves are graphical representations of the various combinations of two goods that provide a consumer with the same level of utility. They are called indifference curves because a consumer is indifferent between any two points on the curve, as they provide the same level of satisfaction.

Mathematically, an indifference curve can be represented by the equation 
$u(x,y) = c$, where u is the utility function, x and y are the quantities of the two goods, and c is a constant representing the level of utility.

An indifference curve is typically downward sloping, meaning that as the quantity of one good increases, the quantity of the other good must decrease to maintain the same level of utility. This is because a consumer faces a tradeoff between the two goods, and as they consume more of one good, they must give up some of the other good to maintain the same level of satisfaction.

#### Budget Line:

The budget line represents the various combinations of two goods that a consumer can afford given their income and the prices of the goods. It is typically represented graphically as a straight line with a slope equal to the ratio of the price of one good to the price of the other good.

Mathematically, the budget line can be represented by the equation $p_xX + p_yY = m$, where $p_x$ and $p_y$ are the prices of the two goods, $X$ and $Y$ are the quantities of the two goods, and m is the consumer's income.

The slope of the budget line represents the relative prices of the two goods. If the price of one good increases relative to the other, the slope of the budget line will become steeper, and the consumer will have to give up more of the other good to afford the same quantity of the more expensive good.

#### Consumer Equilibrium:

Consumer equilibrium occurs when a consumer chooses the combination of two goods that maximizes their utility subject to their budget constraint. This occurs where the budget line is tangent to an indifference curve, as shown in the diagram below.




```python
import numpy as np
import matplotlib.pyplot as plt

# Define the utility function
def utility(x, y):
    return np.sqrt(x*y)

# Define the budget constraint function
def budget(x, px, py, m):
    return (m - px*x) / py

# Define the marginal utility function for good X
def mu_x(x, y):
    return 0.5 * y**(0.5) * x**(-0.5)

# Define the marginal utility function for good Y
def mu_y(x, y):
    return 0.5 * x**(0.5) * y**(-0.5)

# Define the grid of x and y values
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)

# Calculate the utility for each combination of x and y
X, Y = np.meshgrid(x, y)
Z = utility(X, Y)

# Create a contour plot of the utility function
plt.figure()
plt.contour(X, Y, Z, levels=10)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Indifference Curves')

# Set the parameters for the budget constraint
px = 1
py = 2
m = 10

# Calculate the budget line
y_budget = budget(x,px, py, m)

# Plot the budget line
plt.plot(x, y_budget, label='Budget Line')

# Calculate the marginal utilities at the consumer equilibrium
x_star = m/(2*px)
y_star = m/(2*py)
mu_x_star = mu_x(x_star, y_star)
mu_y_star = mu_y(x_star, y_star)

u_level = m / (2 * (px * py) ** 0.5)
x_range = np.linspace(1.25, 10, 50)
y_range = u_level**2 / x_range
plt.plot(x_range, y_range, label='Indifference curve')

# Plot the consumer equilibrium
plt.plot(x_star, y_star, 'ro', label='Consumer Equilibrium')


plt.legend()
plt.show()
```


At the point of tangency, the slope of the budget line $(p_x/p_y)$ is equal to the slope of the indifference curve $(-MU_x/MU_y)$, where $MU_x$ and $MU_y$ are the marginal utilities of X and Y, respectively. This is known as the **marginal rate of substitution (MRS)**, which represents the rate at which a consumer is willing to give up one good to obtain more of the other good while maintaining the same level of satisfaction.

Mathematically, we can express consumer equilibrium as follows:

$$max \quad u(X,Y) \quad s.t. \quad p_xX + p_yY = m$$

The first-order conditions for this problem are:

$$MU_x/p_x = MU_y/p_y$$

These conditions are equivalent to the slope of the budget line being equal to the slope of the indifference curve at the point of tangency.

Let $X$ be the quantity of product X consumed, $Y$ be the quantity of product Y consumed, $U$ be the utility derived from consuming $X$ and $Y$, $m$ be the budget available for consumption, $p_x$ be the price of X, and $p_y$ be the price of Y.

$$max \quad u(X,Y) \quad s.t. \quad p_xX + p_yY = m$$


The Lagrangian function is:

$$L(X,Y,\lambda) = \sqrt{XY} + \lambda (m - p_x X - p_y Y)$$

Taking the partial derivatives with respect to $X$, $Y$, and $\lambda$ and setting them to zero, we get:

$$\frac{\partial L}{\partial X} = \frac{1}{2\sqrt{XY}} Y - \lambda p_x = 0$$

$$\frac{\partial L}{\partial Y} = \frac{1}{2\sqrt{XY}} X - \lambda p_y = 0$$

$$\frac{\partial L}{\partial \lambda} = m - p_x X - p_y Y = 0$$

Solving for $\lambda$ from the first equation, we get:

$$\lambda= \frac{\sqrt{Y}} {2 p_x \sqrt{X}}$$

Solving for $\lambda$ from the second equation, we get:

$$\lambda= \frac{\sqrt{X}} {2 p_y \sqrt{Y}}$$

So, 
$$\frac{\sqrt{Y}} {2 p_x \sqrt{X}}= \frac{\sqrt{X}} {2 p_y \sqrt{Y}}$$

$$Y = \frac{p_x X} {p_y}$$

Substituting this into the budget constraint and solving for $X$, we get:

$$X^* = \frac{m}{2P_x} $$

Substituting $X$ into the first equation above, we get:

$$Y^* = \frac{m}{2P_y} $$

The corresponding maximum utility is:

$$U^* = \sqrt{X^*Y^*} = \sqrt{\frac{m} {2p_x} \frac{m}{2p_y}} = \frac{m}{2\sqrt{p_x p_y}}$$

Therefore, the optimal consumption bundle that maximizes utility subject to the budget constraint of 1000 and $P_x$ of 10 and $P_y$ of 5  is:

$$
\begin{aligned}
X^* = \frac{1000}{2 \cdot 10}
&= 50
\end{aligned}
$$

$$
\begin{aligned}
Y^* = \frac{1000}{2 \cdot 5}
&= 100
\end{aligned}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the budget constraint
def budget(x, m, px, py):
    return m/py - px/py * x

# Define the utility function
def utility(x, y, px, py):
    return (x * y) ** 0.5

# Define the parameters
px = 10
py = 5
m = 1000

# Plot the budget line
x_range = np.linspace(0, m/px, 100)
y_range = budget(x_range, m, px, py)
plt.plot(x_range, y_range, label='Budget line')

# Plot the indifference curve
u_level = m / (2 * (px * py) ** 0.5)
x_range = np.linspace(30, 70, 100)
y_range = u_level**2 / x_range
plt.plot(x_range, y_range, label='Indifference curve')
plt.axvline(x=50, color='gray', linestyle='--')
plt.axhline(y=100, color='gray', linestyle='--')
plt.plot(50, 100, marker='o', color='red', label='Maximum utility')

plt.xlabel('Quantity of X')
plt.ylabel('Quantity of Y')
plt.title('Budget line and indifference curve')
plt.legend()

plt.show()
```
