## Consumer Utility Example

Suppose utility $ u(x,y) = \sqrt{xy} \,$ . Consumer has 1000 dollars to spend.  Price of x is 10 dollars. Price of y is 4 dollars.

* Graph the utility curve in 3-D.
* Find the point that maximizes consumer utility.
* Graph the isoquant through this point on a 2-D graph.
* Calculate expenditure shares.

If the price of x increases by 10 dollars,
* Find the new point that maximizes consumer utility.
* Calculate the income efffect.
* Calculate the substituion effect.


```python
import matplotlib
from matplotlib import animation
import numpy as np
import matplotlib.pyplot as plt

# define utility function
def get_utility(x,y):
    return np.sqrt(x*y)

fig = plt.figure(figsize=(12,8))  #set the default figure size
ax = fig.add_subplot(1,1,1, projection='3d') # add 3rd dimension to subplots

x = np.linspace(0,100,1000)
y = np.linspace(0,100,1000)
X, Y = np.meshgrid(x,y) # create a numpy 2d array
Z = get_utility(X,Y)

# graph a 3d surface plot
ax.plot([0,0,0], [100,100,100])
ax.plot_surface(X,Y,Z, alpha=0.75, color='lightblue')
ax.set_facecolor('y')
fig.set_facecolor('y')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('utility')
plt.title('Utility Function')

# graph a contour plot on z direction to show isoquants
cset = ax.contour(X,Y,Z, zdir='z', offset=0, cmap=matplotlib.cm.coolwarm)
#ax.set_xlim3d(0,60)

ax.view_init(30,-60)

```

```python
# optional animation

#for angle in range(0,360):
#   ax.view_init(0, angle)
#   plt.draw()
#   plt.pause(1)    
#plt.show()

from matplotlib.animation import FuncAnimation

def change_plot(frame):
    plot = ax.plot_surface(X,Y,Z, alpha=0.75, color='lightblue')
    ax.view_init(30, 300-frame*2)
    plot.remove()

animation = FuncAnimation(fig, change_plot, frames = 45, interval=1000)
animation.save('consumer_utility.gif',writer='imagemagick', fps=10)
plt.show()
```
