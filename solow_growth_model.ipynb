### Solow Growth Model

The Solow growth model, also known as the Solow-Swan growth model, is a framework used to understand economic growth in the long run. It was developed independently by Robert Solow and Trevor Swan in the 1950s and 1960s.

The basic idea of the Solow growth model is that there are three factors that contribute to economic growth: **capital accumulation, labor force growth, and technological progress**. The model assumes that these factors interact with each other in a way that determines the long-run rate of economic growth.

The model starts with the assumption that there is a production function that relates the inputs of capital and labor to the output of goods and services. In other words, there is a relationship between the amount of capital ($K$) and labor ($L$) used in production and the amount of output ($Y$) that is produced. The production function is typically assumed to have **diminishing returns to capital and labor**, which means that as more capital and labor are added, the additional output produced from each additional unit of capital or labor decreases.

The model also assumes that there is a constant rate of depreciation of capital ($d$), which means that some amount of capital is lost over time due to wear and tear. The model assumes that there is a constant savings rate ($s$), which determines how much of the economy's output is saved and invested in new capital.

The model then uses these assumptions to derive an equation that shows the long-run rate of economic growth, which is determined by the rate of technological progress ($A$) and the rate of population growth ($n$). The model shows that in the long run, the economy will converge to a steady-state level of capital and output, where the rate of capital accumulation is equal to the aggregate of the rates of depreciation, population growth and technological advancement.

The steady-state expression can be derived from the production function (We will omit technology and technological advancement to simplify the model),

$$\begin{equation*}
Y = F(K, L)
\end{equation*}$$

The Solow growth model is based on the assumption of a production function with constant returns to scale, which is often considered realistic and simplifies the analysis. In a production function with constant returns to scale, any positive multiplication factor ($z$) applied to both capital and labor results in the same multiplication factor applied to output 

$$\begin{equation*}
zY = F(zK, zL)
\end{equation*}$$


This allows for the analysis of all quantities in the economy relative to the size of the labor force. By setting $z = 1/L$ in the above equation, 

$$\begin{equation*}
\frac{Y} {L} = F(K/L, 1)
\end{equation*}$$

We define $y = Y/L$ as output per worker and
$k = K/L$ as capital per worker.
Then, we have 
$$\begin{equation*}
y = f(k),
\end{equation*}$$
where $f(k) = F(k, 1)$.
The marginal product of capital (MPK) can be expressed as 
$$\begin{equation*}
MPK = f(k + 1) âˆ’ f(k)
\end{equation*}$$
Next, we can use the national income accounting identity, 
$$y = c + i,$$
where $y$ is total output or income in the economy, $c$ is consumption expenditure, and $i$ is investment expenditure.
It shows that total output is either consumed or invested in the economy. We can then use the consumption function, 
$$c = (1-s) \cdot y,$$

where $c$ is consumption expenditure, $y$ is total output or income, and $s$ is the savings rate.
It shows that consumption expenditure is a function of total output and the savings rate. The higher the savings rate, the lower the consumption expenditure, and vice versa. Finally, we can rearrange the national income accounting identity to get 
$$y = (1-s) \cdot y + i,$$
which shows that total output or income in the economy is either consumed or invested.

The investment function, 
$$i = s \cdot,$$
represents the relationship between investment expenditure ($i$) and the savings rate ($s$). It shows that investment expenditure is a function of the savings rate. The higher the savings rate, the higher the investment expenditure, and vice versa. In the Solow model, we can modify the investment function to include the level of capital per worker, $i = s \cdot f(k)$,
where $f(k)$ is the production function.
This shows that investment expenditure is a function of the savings rate and the level of capital per worker. The higher the savings rate or the level of capital per worker, the higher the investment expenditure.
Finally, we can express the change in capital per worker ($\Delta k$) as a function of the savings rate ($s$), the production function ($f(k)$), the depreciation rate ($d$), and the population growth rate ($n$) as follows: $$\Delta k = s \cdot f(k) - (d + n) \cdot k$$

At the steady state level,
$$\Delta k = 0$$

So, we can set
$$ s \cdot f(k^*) = (d+n) \cdot k^*$$
to find the steady state level of $k^*$,
$$k^* = s \cdot  \frac {f(k^*)} {(d+n)}$$
This equation shows that the steady state level of capital per worker depends on the parameters of the production function, the savings rate, the depreciation rate, and the population growth rate.

If we assume a Cobb-Douglas form of production function, 
$$Y = A \cdot K^\alpha \cdot L^{1-\alpha}$$

If we assume a Cobb-Douglas form of production function, 
$$Y = A K^{\alpha} L^{1-\alpha},$$
then we can rewrite it in terms of output per worker, 
$$y = \frac{Y} {L},$$ 
as $$y = A k^{\alpha},$$ where $k = K/L$ is the capital per worker.

At the steady state level of capital, we have 
$$k^{\ast} = s \cdot A  \cdot \frac{k^{\ast \alpha}} {(d+n)},$$ 
where $s$ is the savings rate, $d$ is the depreciation rate, and $n$ is the rate of population growth. Rearranging this expression, we get 
$$(k^{\ast})^{(1-\alpha)} = s \cdot \frac{A} {(d+n)},$$ 
which implies that 
$$k^{\ast} = s \cdot  \frac{A} {(d+n)}^{\frac {1} {(1-\alpha)}}$$

To find the steady state level of output per worker ($y^{\ast}$), we substitute $k^{\ast}$ into the production function $y = f(k)$: 
$$y^{\ast} = A(k^{\ast})^{\alpha}$$
This equation shows that the steady state level of output per worker depends on the steady state level of capital per worker ($k^{\ast}$) and the production function ($f$). It also suggests that increasing the level of capital per worker or improving the production function can lead to higher output per worker in the steady state.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
s = 0.2 # Savings rate
n = 0.02 # Labor force growth rate
g = 0.02 # Technological progress rate
delta = 0.05 # Depreciation rate
alpha = 0.3 # Capital share of income
A = 1 # Technology level

# Define the production function
def production_function(K, L):
    return A * K**alpha * L**(1-alpha)

# Define the capital accumulation equation
def capital_accumulation(K, L):
    return s * production_function(K, L) - delta * K

# Define the per-worker production function
def per_worker_production_function(k):
    return A * k**alpha

# Define the per-worker capital accumulation equation
def per_worker_capital_accumulation(k):
    return s * per_worker_production_function(k) - (n+delta) * k

# Find the steady-state level of capital per worker
k_star = ((s*A)/(n+delta))**(1/(1-alpha))

# Find the steady-state level of output per worker
y_star = per_worker_production_function(k_star)

# Define the range of capital per worker values to plot
k = np.linspace(0, 5, 1000)

# Create a figure with size (12, 9) inches
fig = plt.figure(figsize=(12, 9))
# Plot the investment curve
plt.plot(k, per_worker_production_function(k)*s, label='Investment Per Worker-- s * f(k)')

# Plot the capital accumulation curve
plt.plot(k, per_worker_capital_accumulation(k), label='Capital Accumulation Per Worker-- s * f(k)')

# Plot the depreciation and population growth times capital per worker curve
plt.plot(k, k * (delta+n), label='(Depreciation + Population Growth Rate) * Capital Per Worker-- (d+n) * k')

# Plot the steady-state capital point
plt.axvline(k_star, linestyle='--', color='k', label='Steady-State Capital-- k_star')

plt.xlabel('Capital per Worker')
plt.ylabel('Values per Worker')
plt.title('Solow Model')
plt.legend()

plt.show()

```

#### Golden Rule Steady State

The golden rule level of capital refers to the level of capital per worker that maximizes consumption per worker in the steady state. Mathematically, we can find the golden rule level of capital by solving for the value of $k$ that maximizes steady state consumption per worker, given by national income account identity.

$$
\begin{align*}
c^* &= y^* - i^* \\
c^* &= f(k^*) - (d+n) \cdot k^*
\end{align*}
$$
where $c$ is consumption per worker, $k$ is capital per worker, $f(k)$ is the production function, and $d$ is the depreciation rate. To maximize consumption per worker, we need to find the level of $k$ that makes the slope of the consumption function equal to zero. The slope of the consumption function is given by:

$$ \frac{dc^âˆ—}{dk^âˆ—} = fâ€²(k^âˆ—)âˆ’(d+n)$$

where $f'(k)$ is the marginal product of capital. Setting this expression equal to zero and solving for $k$,

$$ 
\begin{aligned}
fâ€²(k_g) \cdot (d+n) &= 0 \\
fâ€²(k_g) &= d+n 
\end{aligned}
$$

This equation shows that the golden rule level of capital occurs where the marginal product of capital equals the depreciation rate. In other words, the optimal level of capital per worker is the one that balances the benefits of having more capital (higher output per worker) with the costs of maintaining that capital (depreciation).

To find the golden rule savings rate, we can use the steady state equilibrium:

$$ 
\begin{aligned}
s_g âˆ— k_g &= (d+n) \cdot k_g \\
s_g &= (d+n) \cdot k_g
\end{aligned}
$$ 

If we assume a Cobb-Douglas form of production function, $Y = A \cdot K^{\alpha} \cdot L^{1-\alpha}$,

$$ 
\begin{aligned}
\frac{Y} {L} &= A \cdot (\frac {K} {L})^{\alpha} \\
y &= A \cdot k^{\alpha}
\end{aligned}
$$

The marginal product of capital (MPK) is:

$$
fâ€²(k) = \frac {dy}{dk} = \alpha \cdot A \cdot k^{\alphaâˆ’1} \\
$$

At the golden rule steady state, 

$$
\begin{aligned}
MPK &= d + n,\\
\alpha \cdot A \cdot k_g^{\alphaâˆ’1} &= d + n
\end{aligned}
$$

The golden rule savings rate is:

$$ s_g = \alpha $$

At the steady-state level of capital,

$$
\begin{align}
fâ€²(k_g) &= d + n \\
\alpha \cdot A \cdot k_g^{\alpha -1} &= d + n \\
k_g &= \bigg( \frac {d+n} {\alpha \cdot A} \bigg)^{\frac {1} {\alpha -1}} 
\end{align}
$$

To find the golden rule level of consumption, we can substitute the expression for the optimal level of $k$ into the consumption function:

$$c_g = f(k_g) âˆ’ (d+n) \cdot k_gâ€‹$$

This expression gives us the maximum level of consumption per worker that can be sustained in the long run, given the economy's production function, depreciation rate, and population growth rate.

```python
# Find the steady-state level of output per worker
y_star = per_worker_production_function(k_star)

# Find the golden rule level of capital per worker
k_gr = ((delta+n)/(alpha * A))**(1/(alpha-1))
y_gr = per_worker_production_function(k_gr)

# Define the range of capital per worker values to plot
k_star = np.linspace(0, 10, 1000)

# Create a figure with size (12, 9) inches
fig = plt.figure(figsize=(12, 9))

# Plot the depreciation and population growth times steady state capital per worker curve
plt.plot(k_star, k_star * (delta+n), label='(Depreciation + Population Growth Rate) * Steady State Capital Per Worker-- (d+n) * k*')

# Plot the Golden Rule Saving Rate times steady state production per worker curve
plt.plot(k_star, per_worker_production_function(k_star)* alpha, label='Golden Rule Savings Rate * Production Per Worker-- s_gold * f(k*)')

# Plot the steady state production per worker curve
plt.plot(k_star, per_worker_production_function(k_star), label='Steady State Production Per Worker-- f(k*)')

# Plot the steady-state capital point
plt.axvline(k_gr, linestyle='--', color='k', label='Golden Rule Steady-State Capital-- k_gold')

# Set x and y axis labels and title
plt.xlabel('Steady State Capital per Worker')
plt.ylabel('Steady State Values per Worker')
plt.title('Solow Model Golden Rule Steady State')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
