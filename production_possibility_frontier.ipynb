## Production Possibility Frontier example

A farmland has 20 workers. By allocating m number of workers they can produce $\sqrt{2 \cdot m}$ tonnes of carrots. By allocating n number of workers they can produce $\sqrt{ 5 \cdot n}$ tonnes of potatoes. Graph the production possibility frontier.
Draw tangent lines and find the values of the slope when the amount of potatoes produced is 1 tonne and 9 tonnes.


### Solution

Let c the kilos of carrots produced. Let p be the kilos of potatoes produced. We have to find the relationship between c an p

Then,

$$
\begin{aligned}
c &= \sqrt{2 \cdot m} \quad => \quad c^2 = 2 \cdot m \quad => \quad \frac {c^2} {2} = m\\
p &= \sqrt{5 \cdot n} \quad => \quad p^2 = 5 \cdot n  \quad => \quad \frac {p^2} {5} = n\\
\\
m + n &= 20
\\
\frac {c^2} {2} + \frac {p^2} {5} &= 20 \\
\frac {c^2} {2} &= 20 - \frac {p^2} {5} \\
c^2 &=  2 \cdot \left( 20 - \frac {p^2} {5} \right) \\
c &= \sqrt{2 \cdot \left( 20 - \frac {p^2} {5} \right)} \\
&= \sqrt{ \left( 40 - \frac {2 \cdot p^2} {5} \right)}
\end{aligned}
$$

Now let's start graphing.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (11,7)  #set the default figure size

# define the function to calculate c for a given value of p
def get_c(p):
    return np.sqrt(40-2*p**2/5)

x = np.linspace(0,9.99,999) # generate numpy 1-D array of points
y = get_c(x) # generate corresponing y values as a numpy 1-D arra

plt.axis([0,10.5,0,7]) # set plot axis dimensions
plt.title("PPF for potatoes and carrots")
plt.xlabel("Tonnes of Potatoes")
plt.ylabel("Tonnes of Carrots")
plt.plot(x,y)
plt.show()

```

We can find the gradient of the function $ c = \sqrt{\left( 40 - \frac {2 \cdot p^2} {5} \right)}$ for $p>=0$ and $c>=0$ manually.

$$
\begin{aligned}
\frac{dc} {dp} &= - \frac {1} {2 \cdot 5} \cdot \frac {4 \cdot p} { \sqrt {\left( 40 - \frac {2 \cdot p^2} {5} \right)}} \\
&= - \frac {2 \cdot p} { 5 \cdot \sqrt {\left( 40 - \frac {2 \cdot p^2} {5} \right)}}
\end{aligned}
$$

Therefore,

$$
\begin{aligned}
\left. \frac{dc} {dp} \right|_{\, p=1}  &= - \frac {2 \cdot 1} { 5 \cdot \sqrt {\left( 40 - \frac {2 \cdot 1^2} {5} \right)}} \\
&\approx - 0.064
\end{aligned}
$$


$$
\begin{aligned}
\\
\left. \frac{dc} {dp} \right|_{\, p=9}  &= - \frac {2 \cdot 9} { 5 \cdot \sqrt {\left( 40 - \frac {2 \cdot 9^2} {5} \right)}} \\
&\approx -1.31
\end{aligned}
$$

Now, we can graph the gradient of the PPF when x=1 and x=9 ( x: the tonnes of potatoes).

```python
# define the fuction to calculate the gradient
def get_c_prime(p): 
    return -2*p /(5 *np.sqrt(40-2*p**2/5))

x = np.linspace(0,9.99,1000) # generate numpy 1-D array of points
y = get_c(x) # generate corresponing y values as a numpy 1-D array

plt.axis([0,10.5,0,7]) # set plot axis dimensions
plt.plot(x,y)
plt.title("PPF for potatoes and carrots")
plt.xlabel("Tonnes of Potatoes")
plt.ylabel("Tonnes of Carrots")

plt.plot(x[100],y[100], marker= 'o', color= 'black', markersize=5) # plot point A1 when x=1
plt.plot(x[900],y[900], marker= 'o', color= 'black', markersize=5) # plot point A2 when x=9

# Graph the gradient m1 when x=1
prime_1 = get_c_prime(1) # claculate the gradient m1 when x=1
x_1 = 1
y_1 = get_c(1) # calculate y value of point A
intercept_1 = y_1 - prime_1 * x_1 # calculate the intercept of the line that goes through point A1 and has gradient m1
x = np.linspace(0.1,1.9, 1000)
y = prime_1 * x + intercept_1  
plt.plot(x,y, color='green')
plt.plot([x[0],x[999]], [y[999], y[999]], linewidth=1, color='orange') # graph a line to show change in X
plt.plot([x[0],x[0]], [y[0], y[999]], linewidth=1, color='purple') # graph a line to show change in y

# Graph the gradient m2 when x=9
prime_2 = get_c_prime(9)
x_2 = 9
y_2 = get_c(9) 
intercept_2 = y_2 - prime_2 * x_2 
x = np.linspace(8.4,9.6, 1000)
y = prime_2 * x + intercept_2
plt.plot(x,y, color='green', label='Gradient (Slope)')
plt.plot([x[0],x[999]], [y[999], y[999]], linewidth=1, color='orange', label=r'$\Delta x$')
plt.plot([x[0],x[0]], [y[0], y[999]], linewidth=1, color='purple', label=r'$\Delta y$')
plt.legend(loc='upper center')

plt.show()
```

Also, we can graph how the gradient changes.

```python
x = np.linspace(0.1,9.9,999)
y = get_c_prime(x)

plt.axis([0,10,-4,0])
plt.plot(x,y)
plt.title("Graph of PPF Gradient")
plt.xlabel("Tonnes of Potatoes")
plt.ylabel("Gradient of PPF")

plt.plot(x[100],y[100], marker= 'o', color= 'black', markersize=5) # plot point A1 when x=1
plt.plot(x[900],y[900], marker= 'o', color= 'black', markersize=5) # plot point A2 when x=9

plt.show()
print("Gradient when tonnes of potatoes = 1 : {gradient:.2f} ".format(gradient=get_c_prime(1)))
print("Gradient when tonnes of potatoes = 9 : {gradient:.2f} ".format(gradient=get_c_prime(9)))
```
