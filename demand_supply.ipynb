## Demand and Supply example

Suppose the demand for gasoline is given by $\quad q_d = 90 - 6 \cdot p \quad$ and the supply for gasoline is given by $\quad q_s = 10 + 2 \cdot p \quad$. Find the equillbrium quantitiy and price. Graph the demand curve and supply curves. Calculate the consumer surplus and supplier surplus.

$\text {At the equillibrium point:}$
$$ \begin{aligned} 
q_d &= q_s \\
90 - 6 \cdot p  &= 10 + 2 \cdot p \quad => \quad -8 \cdot p = -80 \quad => \quad p = 10 \\

\text{Substituting} \; p=10 \; \text{ to:} \quad q_d &= 90 -6 \cdot p \hspace{10cm} \\
q_{eq} &= 90 - 6 \cdot 10  \quad => \quad q_{eq} = 30 
\end {aligned}
$$

$\text{Let's start graphing.}$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,6)  #set the default figure size

# graph the demand curve
# define demand price as a function of quantity
# q_d = 90 -6p => p = 15 - q_d/6
def get_pd(q):
    return 15 -q/6

plt.axis([0,50,0,20])
plt.title("Demand and Supply for Gasoline")
plt.xlabel("quantity")
plt.ylabel("price")
x1 = np.linspace(0,50,1000)
y1 = get_pd(x1)
plt.plot(x1,y1, label='demand', color ='red')

# graph the supply curve
# define supply price as a function of quantity
# q_s = 10 + 2p => p = -5 + q_s/2
def get_ps(q):
    return -5 + q/2

x2 = np.linspace(0,50,1000)
y2 = get_ps(x2)
plt.plot(x2,y2, label='supply', color='blue')

# graph the consumer surplus
x = np.linspace(0,30,1000)
yp = np.linspace(10,10,1000)
plt.plot(x,yp, linestyle='dashed', color='black')
y3 = get_pd(np.linspace(0,30,1000))
plt.fill_between(x,yp,y3, label='consumer surplus', color='pink')

# graph the supplier surplus
x = np.linspace(10,30,1000)
y4 = get_ps(np.linspace(10,30,1000))
plt.fill_between(x,yp,y4, label='supplier surplus', color='lightblue')


plt.legend(loc='upper center')
plt.show


```

To calculate the consumer surplus, we have to find the area between the demand curve $\quad  p_d = 15 - \frac{q_d}{6} \quad $ above the equilbrium price of 10 and between the quantity values of [0, equillibrium quantity]. Equilibrium price is 30. 

$$ 
\begin{aligned}
\int_{0}^{30} \left( 15 - \frac {x} {6} \right) - 10 \quad dx  \quad &= \quad
\int_{0}^{30} 5 - \frac {x} {6} \quad dx \\
\\ 
&= \left. \left[ 5 \cdot x - \frac {x^2} {12} \right] \right|_{0}^{30}
\\
\\
&= 75
\end{aligned}

$$

To calculate the supplier surplus, we have to find the area between the supply curve $ \quad p_s = -5 + \frac{q_s} {2} \quad $ below the equilibrium price of 10 and between the quantity values of [10, equilibrium quantity]. Equilibrium quantity is 30.

$$ 
\begin{aligned}
\int_{10}^{30} 10 - \left( -5 + \frac{x}{2} \right) \quad dx \quad 
&= \int_{10}^{30} 15 - \frac{x}{2} \quad dx \\
\\
&= \left. \left[ 15 \cdot x - \frac {x^2} {4} \right] \right|_{10}^{30} \\
\\
&= 100

\end{aligned}
$$

Optionally, we can use the sympy package to calculate the definite integrals.

```python
import sympy as sym
sym.init_printing(use_unicode=False, wrap_line=False)

x = sym.symbols('x')

f_d = 5 - x/6
f_s = 15 - x/2

#print(sym.integrate(f_d))
#print(sym.integrate(f_s))
print("Consumer surplus is: {0}".format(sym.integrate(f_d, (x,0,30))))
print("Producer surplus is: {0}".format(sym.integrate(f_s, (x,10,30))))
```
